```python
import pandas as pd
```

```python
data = {
    'Age': [22, 25, 30, 35, 40, 45, 50],
    'Marketing_Spend': [2000, 3000, 4000, 5000, 6000, 7000, 8000],
    'Sales': [18000, 24000, 31000, 36000, 42000, 48000, 52000],
    'Product_Category': ['A', 'A', 'B', 'B', 'C', 'C', 'C']
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
#Scatter plots (relationship between two numerical variables)
```

```python
#Marketing Spend vs Sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.scatterplot(
    x='Marketing_Spend',
    y='Sales',
    data=df
)
plt.title('Marketing Spend vs Sales')
plt.show()
```

```python
#Age vs Sales (adding a categorical dimension)
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(
    x='Age',
    y='Sales',
    hue='Product_Category',
    data=df
)
plt.title('Age vs Sales by Product Category')
plt.show()
```

```python
#Compute correlation
```

```python
correlation_matrix = df[['Age', 'Marketing_Spend', 'Sales']].corr()
print(correlation_matrix)
```

```python
#Heatmap visualization
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Pair plot (multiple relationships at once)
```

```python
sns.pairplot(
    df,
    vars=['Age', 'Marketing_Spend', 'Sales'],
    hue='Product_Category'
)
plt.show()
```
