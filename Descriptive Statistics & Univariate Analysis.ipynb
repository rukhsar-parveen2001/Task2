```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create the dataset
```

```python
data = {
    'Age': [25, 30, 35, 40, 28],
    'Income': [40000, 50000, 60000, 80000, 45000],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'Purchase': ['Yes', 'No', 'Yes', 'Yes', 'No']
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
#Identify numerical and categorical columns
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64'])
categorical_cols = df.select_dtypes(include=['object','string'])
```

```python
#Summary statistics for numerical variables
```

```python
numerical_summary = numerical_cols.describe()
print(numerical_summary)
```

```python
#Individual statistics
```

```python
for col in numerical_cols.columns:
    print(f"\nSummary for {col}")
    print("Mean:", numerical_cols[col].mean())
    print("Median:", numerical_cols[col].median())
    print("Standard Deviation:", numerical_cols[col].std())
    print("Min:", numerical_cols[col].min())
    print("Max:", numerical_cols[col].max())
```

```python
#Summary statistics for categorical variables
```

```python
for col in categorical_cols.columns:
    print(f"\nFrequency count for {col}")
    print(df[col].value_counts())
```

```python
#Percentage distribution
```

```python
for col in categorical_cols.columns:
    print(f"\nPercentage distribution for {col}")
    print(df[col].value_counts(normalize=True) * 100)
```

```python
#Visualizations for numerical variables (Histograms)
```

```python
for col in numerical_cols.columns:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=5, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Visualizations for categorical variables (Bar charts)
```

```python
for col in categorical_cols.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
#Key insights (example)
```

```python
print("Average age:", df['Age'].mean())
print("Most common gender:", df['Gender'].mode()[0])
print("Purchase rate (%):", (df['Purchase'].value_counts(normalize=True)['Yes'] * 100))
```
